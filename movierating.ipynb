```python
datamov.Genre.cat.categories
```

```python
list=[]
for gen in datamov.Genre.cat.categories:
  list.append(datamov[datamov.Genre==gen].Budget)
h=plt.hist(list,bins=30,stacked=True)
```

```python
list=[]
label=[]
for gen in datamov.Genre.cat.categories:
  list.append(datamov[datamov.Genre==gen].Budget)
  label.append(gen)
h=plt.hist(list,bins=30,stacked=True,label=label)
plt.legend()
```

```python
vis=sns.lmplot(data=datamov,x='Ratings',y='AudiRatings')
# line in graph is regression line - above line = above avg , below line= low than avg 
```

```python
vis=sns.lmplot(data=datamov,x='Ratings',y='AudiRatings',hue='Genre',fit_reg=False,size=10,aspect=1)
```

```python
sns.kdeplot(datamov.Ratings,datamov.AudiRatings,shade=True, cmap='Reds')
```

```python
#subplot 
f,axes=plt.subplots(2,2,figsize=(12,6))
k1=sns.kdeplot(datamov.Budget,datamov.AudiRatings,shade=True,ax=axes[0,0])
k2=sns.kdeplot(datamov.Budget,datamov.Ratings,shade=True,cmap='Reds',ax=axes[0,1])
k3=sns.kdeplot(datamov.Budget,datamov.AudiRatings,shade=True,cmap='Reds',ax=axes[1,0])
k4=sns.kdeplot(datamov.Budget,datamov.Ratings,shade=True,ax=axes[1,1])
```
